```python
import pandas as pd
import numpy as np
```

```python
campaign = pd.read_csv('data/AMSCampaign.csv')
campaign_details = pd.read_csv('data/AMSCampaignDetail.csv')
campaign_details = campaign_details.join(campaign.set_index('id'), on='campaign')
```

## Campaign with non-zero result

```python
campaign_details
```

```python
campaign_results = campaign_details.groupby(campaign_details.campaign).sum()
campaign_results
```

```python
campaign_results = campaign_results.reset_index()
campaign_results
```

```python
empty_campaigns = campaign_results[campaign_results.impressions ==0].campaign.values
empty_campaigns
```

```python
for campaign in empty_campaigns:
    campaign_results = campaign_results[campaign_results.campaign != campaign]
```

```python
campaign_results.campaign.values
```

```python
campaign_results.T
```

```python
campaign_results = campaign_results[['campaign', 'impressions', 'clics', 'spend', 'sales', 'orders', 'units']]
campaign_results = campaign_results.join(campaign.set_index('id'), on='campaign')
campaign_results = campaign_results.drop(columns = ['currency', 'end_date'])
campaign_results
```

```python
for camp in empty_campaigns:
    campaign_details = campaign_details[campaign_details.campaign != camp]
```

# Campaigns EDA

## Campaigns type - SB, SD, SP

```python
results_by_type = campaign_results.groupby(campaign_results.type)[
    'impressions', 
    'clics', 
    'spend', 
    'sales', 
    'orders', 
    'units'].sum().reset_index()
results_by_type['ROI'] = results_by_type.sales / results_by_type.spend
results_by_type.sort_values(by = 'ROI', ascending = False)
```

## For SP, targeting type

```python
results_by_targeting = campaign_results.groupby(campaign_results.targeting_type)[
    'impressions', 
    'clics', 
    'spend', 
    'sales', 
    'orders', 
    'units'].sum().reset_index()
```

```python
results_by_targeting['ROI'] = results_by_targeting.sales / results_by_targeting.spend
results_by_targeting.sort_values(by = 'ROI', ascending = False)
```

## For SP, daily budget

```python
results_by_daily_budget = campaign_results.groupby(campaign_results.daily_budget)[
    'impressions', 
    'clics', 
    'spend', 
    'sales', 
    'orders', 
    'units'].sum().reset_index()
```

```python
results_by_daily_budget['ROI'] = results_by_daily_budget.sales / results_by_daily_budget.spend
results_by_daily_budget.sort_values(by = 'ROI', ascending = False)
```
