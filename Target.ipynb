```python
import pandas as pd
```

```python
target = pd.read_csv('data/AMSTarget.csv')
target_placement = pd.read_csv('data/AMSTargetPlacementDetail.csv')
placement = pd.read_csv('data/AMSPlacement.csv')
```

```python
placement.campaign1.unique()
```

```python
topic = target_placement[['id', 'target']].join(target.set_index('id'), on='target')
```

```python
topic['placement'] = target_placement[['placement']]
```

```python
placement = placement.rename(columns= {"campaign": "campaign1", "placement": "placement_type"})
placement.placement_type.unique()
```

```python
topic = topic.join(placement.set_index('id'), on = 'placement')
```

```python
topic[['date', 
      'impressions',
      'clicks',
      'CtR',
      'CpC',
      'spend',
      'ACoS']] = (
    target_placement[[
        'date', 
      'impressions',
      'clicks',
      'CtR',
      'CpC',
      'spend',
      'ACoS']])
```

```python
topic[['RoAS', 
      'sales',
      'orders',
      'units',
      'conversion_rate',
      'page_views']] = (
    target_placement[[
        'RoAS', 
      'sales',
      'orders',
      'units',
      'conversion_rate',
      'page_views']])
```
